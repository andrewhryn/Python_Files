```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.to_csv('data_jobs_export.csv', index=False)
```

```python
df_DC_jobs = df[df['job_title_short']=='Data Scientist']
```

```python
df_DC_jobs.to_csv('data_scientist_jobs.csv', index=False)
```

```python
df_NY_jobs = df[df['job_location']=='New York, NY']

df_NY_jobs
```
