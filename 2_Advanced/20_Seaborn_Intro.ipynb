```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df
```

```python
df_DS_US = df[(df['job_title_short']=='Data Scientist') & (df['job_country']=='United States')]
```

```python
top_5 = df_DS_US.groupby('company_name')['salary_year_avg'].median().sort_values(ascending=False).head(5).reset_index()
```

```python
top_5
```

```python
top_5
```

```python
sns.barplot(data=top_5, x = 'salary_year_avg', y='company_name',palette='rocket')

plt.xlabel('Median Salary in USD')
plt.ylabel('')
plt.title('Top 5 hight median salary in USA')
```

```python
df_DS_US = df_DS_US.dropna(subset='salary_year_avg')
```

```python
sns.histplot(df_DS_US['salary_year_avg'], bins=25, kde=True)
plt.xlim(0,700000)
plt.title('Distribution of Yearly Salaries for Data Scientist in the United States')
plt.xlabel('Yearly Salary in USD')
plt.ylabel('Number of Jobs')
```
