```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie', startangle=90, autopct='%1.1f%%')

plt.ylabel('')
plt.title('Work from Home Status')
```

```python
df['job_title_short'].value_counts().plot(kind='pie')

plt.ylabel('')
plt.title('Work from Home Status')
```

```python
fig, ax = plt.subplots()

ax.pie(df['job_work_from_home'].value_counts())
```

```python
df_DA = df[df['job_title_short']=='Data Analyst']
```

```python
df_DA_ins = df_DA['job_health_insurance'].value_counts()

df_DA_ins
```

```python
plt.pie(df_DA_ins,labels=['No', 'Yes'], startangle=90, autopct='%1.1f%%')
plt.title('Proportion of Data Analyst Jobs Mentioning Health Insurance')
```

```python


all = df['job_schedule_type'].value_counts()
```

```python
df_ft_pt = all[['Full-time', 'Part-time']]
```

```python
df_ft_pt
```

```python
plt.pie(df_ft_pt, labels=df_ft_pt.index, startangle=90, autopct='%1.1f%%')
plt.title('Job Schedule Full-time vs. Part-time')
```

```python
df_US = df[df['job_country']=='United States']

df_US
```

```python
df_US_degree = df_US['job_no_degree_mention'].value_counts()

df_US_degree
```

```python
plt.pie(df_US_degree, labels=['Required', 'Not required'], explode=(0, 0.1), autopct='%1.1f%%')

plt.title('Distribution of Jobs Mentioning Degrees for US Job Postings')
```
